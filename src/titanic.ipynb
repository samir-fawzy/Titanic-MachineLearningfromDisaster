```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df = pd.read_csv("data/train.csv")
df.head()
```

```python
df_cls = df.drop(["PassengerId","Name","Ticket","Cabin","SibSp","Parch","Embarked"],axis=1,errors="ignore")
df_cls.head()
```

```python
df_cls["Sex"] = df_cls["Sex"].astype("category").cat.codes
df_cls.head()
```

```python
df_cls.info()
df_cls.describe()
```

```python
df_cls.isnull().sum()
```

```python
df_cls["Age"] = df_cls["Age"].fillna(df_cls["Age"].mean()).astype("int")
```

```python
df_cls[df_cls["Age"].isnull()]
```

```python
df_rescaled = df_cls
df_rescaled.head()
```

```python
for i in range(df_rescaled.shape[1]):
    mean = df_rescaled.iloc[:, i].mean()
    max_ = df_rescaled.iloc[:, i].max()
    min_ = df_rescaled.iloc[:, i].min()
    if max_ != min_:
        df_rescaled.iloc[:,i] = (df_rescaled.iloc[:,i] - mean) / (max_ - min_)
    else:
        df_rescaled.iloc[:,i]= 0
df_rescaled.head()
```

```python
X = df_cls.iloc[:,1:]
Y = df_cls.iloc[:,0]

x = X.to_numpy()
y = Y.to_numpy()
print("=" * 50)
print(x[:10])
print("=" * 50)
print(y[:10])
print(x.shape)
print(y.shape)
print("=" * 50)
```

```python
import tensorflow as tf
y = (y >= 0.5).astype(int)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=4, activation="relu"),
    tf.keras.layers.Dense(units=1, activation="sigmoid")
])

model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(from_logits=False),
    optimizer=tf.keras.optimizers.Adam(),
    metrics=["accuracy"]
)

model.fit(x, y, epochs=100)
```
